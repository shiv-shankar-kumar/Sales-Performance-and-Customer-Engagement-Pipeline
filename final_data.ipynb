```python
# Step 1: Create 'sales_month' column in YYYY-MM format
s3_customer_store_sales_df_join["sales_month"] = pd.to_datetime(
    s3_customer_store_sales_df_join["sales_date"]
).dt.to_period("M").astype(str)

# Step 2: Select the required columns to form final_sales_team_data_mart_df
final_sales_team_data_mart_df = s3_customer_store_sales_df_join[
    [
        "store_id",
        "sales_person_id",
        "sales_first_name",
        "sales_last_name",
        "store_manager_name",
        "manager_id",
        "is_manager",
        "sales_address",
        "sales_pincode",
        "sales_date",
        "total_cost",
        "sales_month"
    ]
]
```

```python
final_sales_team_data_mart_df.head()
```

```python


# Step 1: Ensure 'sales_month' column exists in YYYY-MM format
final_sales_team_data_mart_df.loc[:, "sales_month"] = pd.to_datetime(
    final_sales_team_data_mart_df["sales_date"]
).dt.to_period("M").astype(str)

# Step 2: Compute total sales per store, salesperson, and month
grouped = final_sales_team_data_mart_df.groupby(
    ["store_id", "sales_person_id", "sales_first_name", "sales_last_name", "sales_month"],
    as_index=False
)["total_cost"].sum()

grouped = grouped.rename(columns={"total_cost": "total_sales_every_month"})

# Step 3: Rank salespeople per store and month by total sales
grouped["rnk"] = grouped.groupby(["store_id", "sales_month"])["total_sales_every_month"]\
                        .rank(method="dense", ascending=False)

# Step 4: Calculate incentive (1% of total_sales for rank 1)
grouped["incentive"] = np.where(
    grouped["rnk"] == 1,
    round(grouped["total_sales_every_month"] * 0.01, 2),
    0.00
)

# Step 5: Create full name and select final columns
grouped["full_name"] = grouped["sales_first_name"] + " " + grouped["sales_last_name"]

final_sales_team_data_mart_table = grouped[[
    "store_id",
    "sales_person_id",
    "full_name",
    "sales_month",
    "total_sales_every_month"
]].rename(columns={"total_sales_every_month": "total_sales"})

final_sales_team_data_mart_table["incentive"] = grouped["incentive"]
```
